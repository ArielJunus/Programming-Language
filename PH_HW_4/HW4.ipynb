```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud

df = pd.read_csv('Salary_Data.csv')
```

```python
print("\nTop 5 Occupations")
df["Job Title"].value_counts().nlargest(5)
```

```python
job_title = df['Job Title'].tolist()
print(type(job_title))
print(job_title)
```

```python
dr_count = {i:job_title.count(i) for i in job_title}
print(dr_count)
```

```python
from collections import Counter

dr_count = {i: job_title.count(i) for i in job_title}

# Use Counter to get the top 20 frequencies
counter_object = Counter(dr_count)
top_50 = dict(counter_object.most_common(50))

print(top_50)
```

```python
mask = np.array(Image.open("trophy.png"))

wordcloud = WordCloud(
    background_color="white", 
    mask = mask,
    contour_color ='steelblue',
)
wordcloud.generate_from_frequencies(top_50)
```

```python
plt.imshow(wordcloud)
plt.axis("off")
plt.figure(figsize=(10,10))
plt.show()
```
