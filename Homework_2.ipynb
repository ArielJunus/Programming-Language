```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_1 = pd.read_csv('data_salary_1.csv')
df_2 = pd.read_csv('data_salary_2.csv')


df_1["Education Level"] = df_1["Education Level"].apply(lambda x:"Master's Degree" if x=="Master's" else x)
df_1["Education Level"] = df_1["Education Level"].apply(lambda x:"Bachelor's Degree" if x=="Bachelor's" else x)
df_1["Education Level"] = df_1["Education Level"].apply(lambda x:"PhD" if x=="phD" else x)

df_2["Education Level"] = df_2["Education Level"].apply(lambda x:"Master's Degree" if x=="Master's" else x)
df_2["Education Level"] = df_2["Education Level"].apply(lambda x:"Bachelor's Degree" if x=="Bachelor's" else x)
df_2["Education Level"] = df_2["Education Level"].apply(lambda x:"PhD" if x=="phD" else x)

df_3 = pd.concat([df_1, df_2])
```

```python
## 5 Questions :

## 1. Does age influence someone salary?
## 2. Does year of experience effect someone salary?
## 3. Is there a different between male and female salary?
## 4. Does someone's education level influence someone salary?
## 5. What is the  most popular occupation? 
```

```python
## Salary trend by age

fig, ax = plt.subplots()

ax.set_title("Salary by age differences")
ax.set_xlabel("Age")
ax.set_ylabel("Salaries ($)")

X = df_3["Age"]
Y = df_3["Salary"]

ax.scatter(X,Y,s=12.5)

print("\nthe older they are the higher their salary is")
```

```python
## Salary trend by the year of experience

fig, ax = plt.subplots()

ax.set_title("Salary by Year of Experience differences")
ax.set_xlabel("Years of Experience")
ax.set_ylabel("Salaries ($)")

X = df_3["Years of Experience"]
Y = df_3["Salary"]

ax.scatter(X,Y, s=12.5)

print("\nthe longer their experience are the higher their salary is")
```

```python
print("\nTotal gender")
df_3["Gender"].value_counts(dropna=True)
```

```python
print("\nSalary mean")
df_3.groupby("Gender")["Salary"].mean()
```

```python
## Salary differences between Male and female

fig, ax = plt.subplots()

X = ['Female', 'Male']
Y = [107888.9986, 121389.8709]

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]//2, y[i], ha = 'center')

ax.set_title("Average Salary by Gender differences")
ax.set_xlabel("Gender")
ax.set_ylabel("Average Salaries ($)")

bar_labels = ['#e75480' , '#6CA0DC']

ax.bar(X,Y, color=bar_labels)
addlabels(X,Y)

print("\nMale earn more than female")
```

```python
## This is so we can sort the mean values from the smallest to the highest

education_salary = df_3.groupby("Education Level")["Salary"].mean()
education_salary_sorted = education_salary.sort_values(ascending=True)
education_salary_sorted
```

```python
## The average salary based on the education level

fig, ax = plt.subplots()

ax.set_title("Education Level Average Salary")
ax.set_xlabel("Education Level")
ax.set_ylabel("Average Salary ($)")

X = ['Highschool', 'Bachelor', 'Master', 'PhD'] 
Y = education_salary_sorted

ax.plot(X,Y,'bo-')
plt.grid()
plt.text('Highschool',36706.694196, '     $36,706.69')
plt.text('Bachelor',95082.908640, '     $95,082.90')
plt.text('Master', 130112.056452, '    $130,112.05')
plt.text('PhD',165651.457999, '$165,651.45  ',horizontalalignment='right')
```

```python
## Top 5 most popular job occupations

print("\nTop 5 Occupations")
df_3["Job Title"].value_counts().nlargest(5)
```

```python
## Find what is the most popular occupations

fig, ax = plt.subplots(figsize=(10,5))

X = ['Software Engineer', 'Data Scientist','Software Engineer Manager','Data Analyst','Senior Project Engineer']
Y = df_3["Job Title"].value_counts().nlargest(5)

ax.set_title("Most Popular Occupations")

bar_color = ['#FF2400' , '#3b75af', '#3b75af', '#3b75af', '#3b75af']

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]//2, y[i], ha = 'center')

bars1 = ax.barh(X,Y,color=bar_color)
ax.bar_label(bars1)
```
